```python
ax = sns.countplot(x='Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)
```

```python
sales
```

```python
sns.barplot(x = 'Gender', y = 'Amount', data = sales)
```

Insights = From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

```python
ax = sns.countplot(x='Age Group', hue = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.columns
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)
```

```python
sales_age
```

```python
sns.barplot(x='Age Group', y = 'Amount', data = sales_age)
```

Insights = From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(5)
```

```python
sales_state
```

```python
sns.barplot(x='State', y = 'Orders', data = sales_state)
```

```python
sales_states = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(5)
```

```python
sales_states
```

```python
sns.barplot(x='State', y = 'Amount', data = sales_states)
```

Insights = From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

```python
ax = sns.countplot(x='Marital_Status', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_states = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(5)
sns.barplot(x='Marital_Status', y = 'Amount', data = sales_states, hue = 'Gender')
```

Insights = From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

```python
ax = sns.countplot(x='Occupation', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
sales_states = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(5)
sns.barplot(x='Occupation', y = 'Amount', data = sales_states)
```

Insights = From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

```python
df.columns
```

```python
ax = sns.countplot(x='Product_Category', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
sales_states = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(5)
sns.barplot(x='Product_Category', y = 'Amount', data = sales_states)
plt.xticks(rotation = 'vertical')
plt.show()
```

Insights = From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
sales_states = df.groupby(['Product_ID'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(5)
sns.barplot(x='Product_ID', y = 'Amount', data = sales_states)
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending = False).plot(kind = 'bar')
```

Conclusion = Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
